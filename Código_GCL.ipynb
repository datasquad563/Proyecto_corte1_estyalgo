Pseudocódigo GCL



*   Mf significa el numero filtrado de datos de la base de datos respecto a unafiltración
*   writeln es la equivalencia a un sout en Java o print en python pero en pseudocódigo


```pseudocode
#Definición de variables

var data : array[0,M) x [0,N) of array [0,k] of char ;
var columns : array[0,N) of array[0,k] of char ;
var df : array[0,M) x [0,N) of array [0,k] of char ;

var tipo, combustible : array[0,k] of char ;
var consumo, precio, maletero : int ;

var df_filtrado : array[0,Mf) x [0,N) of array [0,k] of char;
var df_scaled : array[0,Mf) x [0,N) of float ;

var usuario_vector  : array[0,3) of float ;
var autos_vector    : array[0,Mf) x [0,3) of float ;   
var sim_scores      : array[0,Mf) of float ;

var i, j, k, m : int ;
var boton : bool ;
var recomendaciones : array[0,R) of array[0,k] of char;

var min_consumo, max_consumo : float;
var min_precio, max_precio : float;
var min_maletero, max_maletero : float;

boton := true;

var temp_row : array[0,N) of array[0,k] of char;
var temp_score : float;
var puntaje : array[0,Mf) of float;

data := [
    ["Toyota", "Corolla", "Sedán", "Gasolina", 15, 20000, 470, "FWD", "Básica"],
    ["Ford", "Ranger", "Pickup", "Diesel", 11, 35000, 900, "4x4", "Intermedia"],
    ["Tesla", "Model 3", "Sedán", "Eléctrico", 0, 40000, 425, "RWD", "Alta"],
    ["Chevrolet", "Spark", "Hatchback", "Gasolina", 18, 12000, 185, "FWD", "Básica"],
    ["Hyundai", "Tucson", "SUV", "Gasolina", 12, 28000, 620, "AWD", "Intermedia"],
    ["Kia", "Rio", "Sedán", "Gasolina", 16, 15000, 390, "FWD", "Básica"],
    ["BMW", "X5", "SUV", "Diesel", 9, 60000, 750, "AWD", "Alta"],
    ["Audi", "A4", "Sedán", "Gasolina", 13, 45000, 480, "AWD", "Alta"],
    ["Nissan", "Leaf", "Hatchback", "Eléctrico", 0, 32000, 380, "FWD", "Intermedia"],
    ["Jeep", "Wrangler", "SUV", "Gasolina", 10, 42000, 1000, "4x4", "Alta"],
    ["Honda", "Civic", "Sedán", "Gasolina", 14, 25000, 450, "FWD", "Intermedia"],
    ["Mazda", "CX-5", "SUV", "Gasolina", 12, 30000, 550, "AWD", "Intermedia"],
    ["Renault", "Duster", "SUV", "Diesel", 14, 21000, 475, "FWD", "Básica"],
    ["Mercedes-Benz", "GLC", "SUV", "Gasolina", 11, 55000, 600, "AWD", "Alta"],
    ["Volkswagen", "Golf", "Hatchback", "Hibrido", 17, 23000, 380, "FWD", "Intermedia"]
];
columns := ["Marca","Modelo","Tipo","Combustible","Consumo","Precio","Maletero","Tracción","Tecnología"];
df := data;


#Inputs de usuario

writeln("Te recomendamos tu auto ideal");

writeln("Seleccione tipo de auto (Sedán / SUV / Pickup / Hatchback): ");
readln(tipo);

writeln("Seleccione tipo de combustible (Gasolina / Diesel / Eléctrico / Híbrido): ");
readln(combustible);

writeln("Ingrese consumo mínimo (km/L): ");
readln(consumo);

writeln("Ingrese precio máximo: ");
readln(precio);

writeln("Ingrese capacidad mínima del maletero (L): ");
readln(maletero);

i := 0;
j := 0;

#Filtrar dataset
do i < M ->
    if df[i,"Tipo"] = tipo ∧ df[i,"Combustible"] = combustible ->
            df_filtrado[j] := df[i];
            j := j + 1
        []  ->
            skip
    fi;
    i := i + 1
od;

#sistema
if boton = true ->
  if j = 0->
      writeln("No hay autos que cumplan esas características.")
  [] ->
      min_consumo := min(df_filtrado[,"Consumo"]);
      max_consumo := max(df_filtrado[,"Consumo"]);
      min_precio := min(df_filtrado[,"Precio"]);
      max_precio := max(df_filtrado[,"Precio"]);
      min_maletero := min(df_filtrado[,"Maletero"]);
      max_maletero := max(df_filtrado[,"Maletero"]);
      
      i := 0;
      do i < j ->
        df_scaled[i, "Consumo"] := df_filtrado[i, "Consumo"];
        df_scaled[i, "Precio"] := df_filtrado[i, "Precio"];
        df_scaled[i, "Maletero"] := df_filtrado[i, "Maletero"];
        i := i+1
      od;

      i:= 0 ;
      #Normalizar columna Consumo
      do i < j ->
          df_scaled[i,"Consumo"] :=
              (df_filtrado[i,"Consumo"] - min_consumo) / (max_consumo - min_consumo);
          i := i + 1
      od;
      #Normalizar columna Precio
      i := 0;
      do i < j ->
          df_scaled[i,"Precio"] :=
              (df_filtrado[i,"Precio"] - min_precio) / (max_precio - min_precio);
          i := i + 1
      od;
      #Normalizar columna Maletero
      i := 0;
      do i < j ->
          df_scaled[i,"Maletero"] :=
              (df_filtrado[i,"Maletero"] - min_maletero) / (max_maletero - min_maletero);
          i := i + 1
      od;

      usuario_vector[0] := (consumo - min_consumo) / (max_consumo - min_consumo);
      usuario_vector[1] := (precio - min_precio) / (max_precio - min_precio);
      usuario_vector[2] := (maletero - min_maletero) / (max_maletero - min_maletero);
      
      i := 0;
      do i<j ->
           autos_vector[i,0] := (df_filtrado[i,"Consumo"] - min_consumo) / (max_consumo - min_consumo);
           autos_vector[i,1] := (df_filtrado[i,"Precio"] - min_precio) / (max_precio - min_precio);
           autos_vector[i,2] := (df_filtrado[i,"Maletero"] - min_maletero) / (max_maletero - min_maletero);
           i := i + 1;
      od;
      #Calculo de similitud
      k := 0;
      do k < j ->
          sim_scores[k] :=
              (usuario_vector[0]*autos_vector[k,0] +
              usuario_vector[1]*autos_vector[k,1] +
              usuario_vector[2]*autos_vector[k,2])
              /
              (sqrt (usuario_vector[0]^2 + usuario_vector[1]^2 + usuario_vector[2]^2)
               * sqrt(autos_vector[k,0]^2 + autos_vector[k,1]^2 + autos_vector[k,2]^2) );
           k := k + 1;                
      od;
      i := 0;
      do i < j ->
          puntaje[i] := sim_scores[i];
          i := i + 1;
      od;
      i := 0;
      do i < j - 1 ->
          k := 0;
          do k < j - i - 1 ->
              if puntaje[k] < puntaje[k+1] ->
                  #Organización de puntajes
                  m := 0;
                  do m< N ->
                    temp_row[m] := df_filtrado[k, m];
                    df_filtrado[k, m] := df_filtrado[k+1, m];
                    df_filtrado[k+1, m] := temp_row[m];
                    m := m + 1;
                  od;  

                  temp_score := puntaje[k];
                  puntaje[k] := puntaje[k+1];
                  puntaje[k+1] := temp_score;
              [] -> skip
              fi;
              k := k + 1;
          od;
          i := i + 1;
       od;
       #Presentacón al usuario
       i := 0;
       do i < j ->
          writeln(df_filtrado[i,"Marca"], " ", df_filtrado[i,"Modelo"], " Puntaje: ", puntaje[i]);
          i := i + 1;
       od;              
  fi;
[] boton = false ->
  skip  
fi;


```
